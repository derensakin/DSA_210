Total Usage Time of Each App

This code calculates and visualizes the total usage time of each app. It is important to determine the most used apps, which screen time and a lot of data can be presented. For example, this information is critical to analyze whether I spent more time on productive apps or entertainment apps during exam week.

Conclusion: By determining the most used apps, I can see how each device is included in the screen time during exam week.

```python
# Her uygulamanın toplam kullanım süresini hesaplıyorum
# 'Date', 'Daily Total Time' ve kategori sütunlarını hariç tutarak sadece uygulamaları seçiyorum
application_columns = data.columns.difference(['Date', 'Daily Total Time', 'Social', 'Entertainment', 'Games',
                                                'Productivity & Finance', 'Creativity', 'Shopping', 'Other', '00.00-06.00', '06.00-12.00', '12.00-18.00', '18.00-00.00'])

# Uygulama sütunlarının toplamını alıyoruz
total_usage_per_app = data[application_columns].sum().sort_values(ascending=False)

# Sonuçları ekrana yazdırıyoruz
print(total_usage_per_app)

# İsteğe bağlı olarak bir bar grafiği oluşturabiliriz
plt.figure(figsize=(15, 6))
total_usage_per_app.plot(kind='bar')
plt.title("Total Usage Time of Each Application")
plt.xlabel("Applications")
plt.ylabel("Total time (min)")
plt.xticks(rotation=45)
plt.show()

```

Calculating and Visualizing Daily Total Screen Time

This code allows me to calculate and visualize my daily total screen time. This metric is especially important for analyzing changes during exam week.

The graph allows me to easily see fluctuations in my daily total screen time and whether usage drops or increases on certain days (e.g. exam days). This is an important step in understanding how my daily routines affect screen time.

Conclusion: This analysis helps me understand how I can optimize screen time or which apps I use less on exam days.

```python
# Günlük toplam kullanım sürelerini hesaplıyorum
daily_total_usage = data.groupby('Date')['Daily Total Time'].sum()

# Sonuçları ekrana yazdırıyorum
print(daily_total_usage)

# Günlük toplam kullanım süresini görselleştiriyorum
plt.figure(figsize=(15, 6))
daily_total_usage.plot(kind='line', marker='o')
plt.title("Total Daily Usage Times")
plt.xlabel("Date")
plt.ylabel("Total time(min)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

Time of Use Analysis by Hours

This code analyzes how my screen time changes over the course of the day. The data shows my screen time intensity during the morning (06:00-12:00) and afternoon (12:00-18:00). This information helps me understand when I am more productive or more playful.

Conclusion: The distribution of my hourly screen time helps me understand my phone usage habits throughout the day and offers tips for making these habits more productive.

```python
# Saat aralıkları sütunlarını belirtiyorum
time_slots = ['00.00-06.00', '06.00-12.00', '12.00-18.00', '18.00-00.00']

# Her saat aralığındaki toplam kullanım süresini hesaplıyorum
total_usage_per_time_slot = data[time_slots].sum()

# Sonuçları ekrana yazdırıyorum
print(total_usage_per_time_slot)

# Saat aralıklarına göre kullanım sürelerini görselleştiriyorum
plt.figure(figsize=(10, 5))
total_usage_per_time_slot.plot(kind='bar', color='orange')
plt.title("Total Usage Time by Hourly Intervals")
plt.xlabel("Hourly Intervals")
plt.ylabel("Total time (min)")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

Blu TV and Netflix Usage Analysis

This code allows us to compare the Blu TV and Netflix applications on exam days and the days after the exam. Camera Usage and Photo Storage Calculation

This code calculates the number of photos taken per day and the storage required for 1 month of photos based on the time spent in the camera app. My camera usage shows that I take an average of 102.86 photos per day.

I assumed that a photo takes up an average of 3MB of space, and based on the number of photos taken per day, I calculated that a total of 9.04GB of storage is required for a month. This information is invaluable for managing my phone's memory and preventing storage issues in the future.

Conclusion: This calculation helped me understand my current camera usage habits and predict my future storage needs.

```python
# Kamera kullanım sürelerini alıyoruz ve bir kopyasını oluşturuyoruz
camera_usage = data[['Date', 'Camera']].copy()

# Fotoğraf çekmek için harcanan süre (her 5 saniyede bir fotoğraf çekiliyor)
photo_time_per_photo = 5  # saniye

# Kamera kullanım süresini dakikadan saniyeye çeviriyoruz
camera_usage['Total Seconds'] = camera_usage['Camera'] * 60

# Günlük çekilen fotoğraf sayısını hesaplıyoruz
camera_usage['Photos Taken'] = camera_usage['Total Seconds'] // photo_time_per_photo

# Günlük ortalama çekilen fotoğraf sayısını hesaplıyoruz
average_photos = camera_usage['Photos Taken'].mean()

# Sonucu yazdırıyoruz
print(f"Average number of photos taken per day: {average_photos:.2f}")
```

```python
# Her fotoğrafın boyutu (MB)
photo_size_mb = 3

# Günlük depolama ihtiyacı (MB)
daily_storage_mb = average_photos * photo_size_mb

# Aylık depolama ihtiyacı (MB) (30 gün üzerinden)
monthly_storage_mb = daily_storage_mb * 30

# MB'yi GB'ye çeviriyoruz
monthly_storage_gb = monthly_storage_mb / 1024

# Sonucu yazdırıyoruz
print(f"For 1 month of usage, you need to free up approximately {monthly_storage_gb:.2f} GB of space in your gallery. ")
```
